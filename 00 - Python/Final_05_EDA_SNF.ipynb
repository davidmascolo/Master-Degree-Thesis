### Custom Functions

```python
def find_characteristics(df, community):
    """Input: clinical table hq, community to analyze
       Output: characterstics of patients"""

    ## Define empty lists
    kpi_01 = []
    kpi_02 = []
    kpi_03 = []
    kpi_04 = []
    kpi_05 = []
    kpi_06 = []
    kpi_07 = []
    kpi_08 = []
    kpi_09 = []
    kpi_10 = []
    kpi_11 = []
    kpi_12 = []
    kpi_13 = []
    kpi_14 = []
    kpi_15 = []
    kpi_16 = []
    kpi_17 = []
    kpi_18 = []
    kpi_19 = []
    kpi_20 = []
    kpi_21 = []
    kpi_22 = []

    ## Columns for which we want the informations
    columns_of_interest = ["Harmonized_SU2C_RNA_Tumor_Sample_ID_v2", "Pre-treatment_RNA_Sample_QC", "Patient_Age_at_Diagnosis", "Patient_Sex", "Patient_Race",
                        "Patient_Smoking_Status", "Patient_Smoking_Pack_Years_Harmonized", "Histology_Harmonized", "Histology_Detail", "Initial_Stage",
                        "PDL1_TPS", "PDL1_TPS_Description", "Local_Antibody_Clone", "Clinical_Driver", "Sequencing_Platform", "Advanced_Diagnosis_Date",
                        "Line_of_Therapy", "Agent_PD1", "Agent_PD1_Category", "Prior_Platinum", "Prior_TKI","Harmonized_Confirmed_BOR_3_Cat", "Harmonized_Confirmed_BOR"]

    ## Filter the columns
    df_filtered = df.loc[:, columns_of_interest]

    ## Cycle
    for i in range(len(community["ID_Patient"])):
        ## Extract patient id
        id_patient = community["ID_Patient"][i]
        ## Filter clinical table for patient id
        filtered_df = df_filtered[df_filtered["Harmonized_SU2C_RNA_Tumor_Sample_ID_v2"] == id_patient]

        ## Extract Characteristics
        ## Pre-treatmant RNA
        kpi_01.append(filtered_df["Pre-treatment_RNA_Sample_QC"].values)
        kpi_01_values = [arr[0] for arr in kpi_01]
        ## Age at diagnosis
        kpi_02.append(filtered_df["Patient_Age_at_Diagnosis"].values)
        kpi_02_values = [arr[0] for arr in kpi_02]
        ## Sex
        kpi_03.append(filtered_df["Patient_Sex"].values)
        kpi_03_values = [arr[0] for arr in kpi_03]
        ## Race
        kpi_04.append(filtered_df["Patient_Race"].values)
        kpi_04_values = [arr[0] for arr in kpi_04]
        ## Smoking Status
        kpi_05.append(filtered_df["Patient_Smoking_Status"].values)
        kpi_05_values = [arr[0] for arr in kpi_05]
        ## Smoking pack years harmonized
        kpi_06.append(filtered_df["Patient_Smoking_Pack_Years_Harmonized"].values)
        kpi_06_values = [arr[0] for arr in kpi_06]
        ## Histology harmonized
        kpi_07.append(filtered_df["Histology_Harmonized"].values)
        kpi_07_values = [arr[0] for arr in kpi_07]
        ## Histology detail
        kpi_08.append(filtered_df["Histology_Detail"].values)
        kpi_08_values = [arr[0] for arr in kpi_08]
        ## Initial stage
        kpi_09.append(filtered_df["Initial_Stage"].values)
        kpi_09_values = [arr[0] for arr in kpi_09]
        ## PDL1
        kpi_10.append(filtered_df["PDL1_TPS"].values)
        kpi_10_values = [arr[0] for arr in kpi_10]
        ## PDL1 Description
        kpi_11.append(filtered_df["PDL1_TPS_Description"].values)
        kpi_11_values = [arr[0] for arr in kpi_11]
        ## Local antibody clone
        kpi_12.append(filtered_df["Local_Antibody_Clone"].values)
        kpi_12_values = [arr[0] for arr in kpi_12]
        ## Clinical driver
        kpi_13.append(filtered_df["Clinical_Driver"].values)
        kpi_13_values = [arr[0] for arr in kpi_13]
        ## Sequencing Platform
        kpi_14.append(filtered_df["Sequencing_Platform"].values)
        kpi_14_values = [arr[0] for arr in kpi_14]
        ## Advanced Diagnosis Date
        kpi_15.append(filtered_df["Advanced_Diagnosis_Date"].values)
        kpi_15_values = [arr[0] for arr in kpi_15]
        ## Line of therapy
        kpi_16.append(filtered_df["Line_of_Therapy"].values)
        kpi_16_values = [arr[0] for arr in kpi_16]
        ## Agent PD1
        kpi_17.append(filtered_df["Agent_PD1"].values)
        kpi_17_values = [arr[0] for arr in kpi_17]
        ## Agent PD1 category
        kpi_18.append(filtered_df["Agent_PD1_Category"].values)
        kpi_18_values = [arr[0] for arr in kpi_18]
        ## Prior Platinum
        kpi_19.append(filtered_df["Prior_Platinum"].values)
        kpi_19_values = [arr[0] for arr in kpi_19]
        ## Prior TKI
        kpi_20.append(filtered_df["Prior_TKI"].values)
        kpi_20_values = [arr[0] for arr in kpi_20]
        ## Harmonized Confirmed BOR Cat
        kpi_21.append(filtered_df["Harmonized_Confirmed_BOR_3_Cat"].values)
        kpi_21_values = [arr[0] for arr in kpi_21]
        ## Harmonized BOR
        kpi_22.append(filtered_df["Harmonized_Confirmed_BOR"].values)
        kpi_22_values = [arr[0] for arr in kpi_22]

    ## Organize into dataframe
    ## Collect data into dictonary
    data = {"Pre-treatment_RNA_Sample_QC": kpi_01_values,
            "Patient_Age_at_Diagnosis": kpi_02_values,
            "Patient_Sex": kpi_03_values,
            "Patient_Race": kpi_04_values,
            "Patient_Smoking_Status": kpi_05_values,
            "Patient_Smoking_Pack_Years_Harmonized": kpi_06_values,
            "Histology_Harmonized": kpi_07_values,
            "Histology_Detail": kpi_08_values,
            "Initial_Stage": kpi_09_values,
            "PDL1_TPS": kpi_10_values,
            "PDL1_TPS_Description": kpi_11_values,
            "Local_Antibody_Clone": kpi_12_values,
            "Clinical_Driver": kpi_13_values,
            "Sequencing_Platform": kpi_14_values,
            "Advanced_Diagnosis_Date": kpi_15_values,
            "Line_of_Therapy": kpi_16_values,
            "Agent_PD1": kpi_17_values,
            "Agent_PD1_Category": kpi_18_values,
            "Prior_Platinum": kpi_19_values,
            "Prior_TKI": kpi_20_values,
            "Harmonized_Confirmed_BOR_3_Cat": kpi_21_values,
            "Harmonized_Confirmed_BOR": kpi_22_values,
            "ID_Patient": id_patient}
    ## Create df
    df_kpi = pd.DataFrame(data)
    ## Remove column
    return(df_kpi)
```

```python
## Create a function that is used to make different plots
def plot_for_class(data_01, data_02, feature, subtitle, col):
    """Input: data, variable that is analyzed subtitle, col"""
    """Output: Conditioned Barplot with percentage for each class"""

    ## For generale features
    ## Compute summary
    summary_01    = data_01.groupby([feature]).size().sort_values(ascending = True).\
                                                                                reset_index(name = "Count").set_index(feature).sort_values(by = "Count")
    summary_02    = data_02.groupby([feature]).size().sort_values(ascending = True).\
                                                                                reset_index(name = "Count").set_index(feature).sort_values(by = "Count")
    # summary_03    = data_03.groupby([feature]).size().sort_values(ascending = True).\
    #                                                                              reset_index(name = "Count").set_index(feature).sort_values(by = "Count")
    ## Percentage
    summary_01["%"] = round(summary_01["Count"] / sum(summary_01["Count"]) * 100, 3)
    display(summary_01)
    print("\n")
    summary_02["%"] = round(summary_02["Count"] / sum(summary_02["Count"]) * 100, 3)
    display(summary_02)
    print("\n")
    # summary_03["%"] = round(summary_03["Count"] / sum(summary_03["Count"]) * 100, 3)
    # display(summary_03)
    # print("\n")
    ## Compute labels
    labels_01    = [f"{value}%" for value in summary_01["%"]]
    labels_02    = [f"{value}%" for value in summary_02["%"]]
    # labels_03    = [f"{value}%" for value in summary_03["%"]]

    ## Institution Feature Case
    if feature in ["Institution"]:
        ## Plot
        plt.figure(figsize = (44, 8))
        plt.subplot(121)
        plt.bar_label(plt.barh(summary_01.index.astype(str), summary_01["Count"], color = col), labels = labels_01)
        plt.title("Community 01")
        plt.suptitle(subtitle)

        plt.subplot(122)
        plt.bar_label(plt.barh(summary_02.index.astype(str), summary_02["Count"], color = col), labels = labels_02)
        plt.title("Community 02")
        plt.suptitle(subtitle)

        # plt.subplot(133)
        # plt.bar_label(plt.barh(summary_03.index.astype(str), summary_03["Count"], color = col), labels = labels_03)
        # plt.title("Community 03")
        # plt.suptitle(subtitle)

        plt.show()
        pass

    ## PD1 Agent Case
    if feature in ["Agent_PD1"]:
        ## Plot
        plt.figure(figsize = (44, 8))
        plt.subplot(121)
        plt.bar_label(plt.barh(summary_01.index.astype(str), summary_01["Count"], color = col), labels = labels_01)
        plt.title("Community 01")
        plt.suptitle(subtitle)

        plt.subplot(122)
        plt.bar_label(plt.barh(summary_02.index.astype(str), summary_02["Count"], color = col), labels = labels_02)
        plt.title("Community 02")
        plt.suptitle(subtitle)

        # plt.subplot(133)
        # plt.bar_label(plt.barh(summary_03.index.astype(str), summary_03["Count"], color = col), labels = labels_03)
        # plt.title("Community 03")
        # plt.suptitle(subtitle)

        plt.show()
        pass

    ## Age Feature Case
    if feature in ["Patient_Age_at_Diagnosis"]:
        ## Divide the age in different range
        patient_age_filtered_02 = data_02["Patient_Age_at_Diagnosis"].dropna()
        patient_age_filtered_02 = patient_age_filtered_02.values.tolist()
        ## Patient_age_filtered_02.remove(">89")
        patient_age_filtered_01 = data_01["Patient_Age_at_Diagnosis"].dropna()
        patient_age_filtered_01 = patient_age_filtered_01.values.tolist()
        patient_age_filtered_01.remove(">89")
        # patient_age_filtered_03 = data_03["Patient_Age_at_Diagnosis"].dropna()
        # patient_age_filtered_03 = patient_age_filtered_03.values.tolist()
        # patient_age_filtered_03.remove(">89")

        ## Convert each element of the list into a int
        patient_age_filtered_01 = list(map(int, patient_age_filtered_01))
        patient_age_filtered_02 = list(map(int, patient_age_filtered_02))
        # patient_age_filtered_03 = list(map(int, patient_age_filtered_03))

        ## Init the dictonary to storage the counts for each set
        # counts_03 = {}
        counts_02 = {}
        counts_01 = {}

        ## Divide the sequence of intervals into 10 and sum the counts for each set
        # for num in patient_age_filtered_03:
        #     interval = num // 10  ## Compute the interval
        #     counts_03[interval] = counts_03.get(interval, 0) + 1  ## Update the interval
        for num in patient_age_filtered_02:
            interval = num // 10  ## Compute the interval
            counts_02[interval] = counts_02.get(interval, 0) + 1  ## Update the interval
        for num in patient_age_filtered_01:
            interval = num // 10  ## Compute the interval
            counts_01[interval] = counts_01.get(interval, 0) + 1  ## Update the interval

        ## Print the counts for each set
        # interval_list_03    = []
        # count_list_03       = []
        interval_list_02    = []
        count_list_02       = []
        interval_list_01    = []
        count_list_01       = []
        # for interval, count in counts_03.items():
        #     interval_list_03.append(f"{interval*10}-{(interval*10)+9}")
        #     count_list_03.append(count)
        for interval, count in counts_02.items():
            interval_list_02.append(f"{interval*10}-{(interval*10)+9}")
            count_list_02.append(count)
        for interval, count in counts_01.items():
            interval_list_01.append(f"{interval*10}-{(interval*10)+9}")
            count_list_01.append(count)
        ## Save dataframe
        summary_age_01 = pd.DataFrame({"Interval": interval_list_01, "Count": count_list_01}).set_index("Interval").sort_values(by = "Interval", ascending = True)
        display(summary_age_01)
        print("\n")
        summary_age_02 = pd.DataFrame({"Interval": interval_list_02, "Count": count_list_02}).set_index("Interval").sort_values(by = "Interval", ascending = True)
        display(summary_age_02)
        print("\n")
        # summary_age_03 = pd.DataFrame({"Interval": interval_list_03, "Count": count_list_03}).set_index("Interval").sort_values(by = "Interval", ascending = True)
        # display(summary_age_03)
        # print("\n")
        ## Percentage
        # summary_age_03["%"] = round(summary_age_03["Count"] / sum(summary_age_03["Count"]) * 100, 3)
        summary_age_02["%"] = round(summary_age_02["Count"] / sum(summary_age_02["Count"]) * 100, 3)
        summary_age_01["%"] = round(summary_age_01["Count"] / sum(summary_age_01["Count"]) * 100, 3)
        ## Compute labels
        # labels_03    = [f"{value}%" for value in summary_age_03["%"]]
        labels_02    = [f"{value}%" for value in summary_age_02["%"]]
        labels_01    = [f"{value}%" for value in summary_age_01["%"]]

        ## Distribution of Age
        plt.figure(figsize = (22, 5))
        plt.subplot(121)
        plt.bar_label(plt.bar(summary_age_01.index.astype(str), summary_age_01["Count"], color = "steelblue"), labels = labels_01)
        plt.title("Community 01")
        plt.suptitle(subtitle)

        plt.subplot(122)
        plt.bar_label(plt.bar(summary_age_02.index.astype(str), summary_age_02["Count"], color = "steelblue"), labels = labels_02)
        plt.title("Community 02")
        plt.suptitle(subtitle)

        # plt.subplot(133)
        # plt.bar_label(plt.bar(summary_age_03.index.astype(str), summary_age_03["Count"], color = "steelblue"), labels = labels_03)
        # plt.title("Community 03")
        # plt.suptitle(subtitle)

        plt.show()
        pass

    if feature not in ["Institution", "Patient_Age_at_Diagnosis", "Agent_PD1"]:
        ## Plot
        plt.figure(figsize = (22, 5))
        plt.subplot(121)
        plt.bar_label(plt.bar(summary_01.index.astype(str), summary_01["Count"], color = col), labels = labels_01)
        plt.title("Community 01")


        ## Plot
        plt.subplot(122)
        plt.bar_label(plt.bar(summary_02.index.astype(str), summary_02["Count"], color = col), labels = labels_02)
        plt.suptitle(subtitle)
        plt.title("Community 02")

        # ## Plot
        # plt.subplot(133)
        # plt.bar_label(plt.bar(summary_03.index.astype(str), summary_03["Count"], color = col), labels = labels_03)
        # plt.suptitle(subtitle)
        # plt.title("Community 03")

        plt.show()
```
